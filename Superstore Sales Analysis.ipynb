```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# **Upload Dataset**

```
df=pd.read_csv("Superstore.csv")
```

# **Data cleaning**

```
df.head()
```

```
df.tail()
```

```
df.describe()
```

```
df.isnull().sum()
```

# **Which Category has highest sales**

```
category_sales=df.groupby('Category')['Sales'].sum()
print(category_sales)
```

```
category_sales.plot(kind='bar')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```

# **Which region has the highest profit**

```
region_profit=df.groupby('Region')['Profit'].sum()
print(region_profit)
```

```
region_profit.plot(kind='bar')
plt.title('Profit by region')
plt.xlabel('Region')
plt.ylabel('Profit')
plt.show()
```

# **Highest Sub-Category has the highest Sales**

```
sub_sales=df.groupby('Sub-Category')['Sales'].sum()
print(sub_sales)
```

```
sub_sales.plot(kind='bar')
plt.title('Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.show()
```

# **Distribution of Sales**

```
plt.hist(df['Sales'],bins=20)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

# **Relation between Sales and Profit**

```
plt.scatter(df['Sales'],df['Profit'])
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

# **Category share (Pie Chart)**

```
category_sales=df.groupby('Category')['Sales'].sum()
plt.pie(category_sales,labels=category_sales.index,autopct='%1.1f%%')
plt.title('Category Sales Share')
plt.show()
```
